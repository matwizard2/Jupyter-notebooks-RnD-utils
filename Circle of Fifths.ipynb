# Circle of Fifths

This code identifies a musical note name (like C, F#, or Bb) based on its position on the Circle of Fifths, relative to C.
Here is the breakdown of how it works:
- Input (position): An integer representing steps away from C on the Circle of Fifths.
- For example, -1 is F, 0 is C, 1 is G, and -2 is Bb.
- Letter Calculation: It uses a mathematical formula—chr((4 * position_from_f + 5) % 7 + 65)—to map the position to the correct letter (A–G).
- Accidentals: It determines how many sharps or flats to add based on how many "loops" around the seven basic notes the position has traveled:
- Sharps (#/x): For positive positions, it uses # for single sharps and x for double sharps.
- Flats (b): For negative positions, it appends the appropriate number of b symbols.
  
Essentially, it’s a programmatic way to navigate musical keys and their associated accidentals.

```python
def note_name(position):
    position_from_f = position + 1
    number_of_accidentals = position_from_f // 7  # 0 = Natural, 1 = Sharp, 2 = Double Sharp etc., -1 = Flat, -2 = Double Flat etc.
    name = chr((4 * position_from_f + 5) % 7 + 65)     # Initialise name as the letter A-G with no sharps or flats.
    
    if position_from_f>= 0:   # If note is natural or sharp (or double, triple sharp etc.) add '#' or 'x' signs
        name += (number_of_accidentals % 2) * '#' + (number_of_accidentals // 2) * 'x'
    else:                     # If note is flat (or double flat etc.) add 'b' signs
        name += -number_of_accidentals*'b'
    return name
```
