# Monty Hall Problem Analyser

This notebook explores the famous Monty Hall game show problem, where the correct approach is counterintuitive to most people.

Monty, the host', asks contestants to choose one of three closed doors, one with a new car and two with goats, and then, afterthe contestant chooses a door, opens and reveals a third door (which is always a goat), and asks if the contesatant wicvhes to chainge her or his choice.

This program simulates and displays the change door and dont change door approaches in terms of winning the car.

[Monty Hall Problem](https://en.wikipedia.org/wiki/Monty_Hall_problem)

```python
def monty(switch=False):
    doors_unopened = {1,2,3}
    car = random.randint(1,3)
    chosen = random.randint(1,3)
    monty_doors = doors_unopened - {car} - {chosen}
    opened = random.choice(tuple(monty_doors))
    doors_unopened -= {opened}

    #print("Car: Door "+str(car))
    #print("Chosen: Door "+str(chosen))
    #print("Door opened by Monty: Door "+str(opened))


    #print('\nIf contestant does not switch: ' + str (chosen==car) + '\n')

    switch_door = random.choice(tuple(doors_unopened - {chosen}))

    #print('Contestant changes guess to: Door '+str(switch_door))

    #print('\nIf contestant switches: ' + str (switch_door==car))
    
    if switch:
        return(switch_door==car)
    else:
        return(chosen==car)
```

```python
win_probability = []
wins = 0
for i in range(1,10001):
    if monty(switch=False):
        wins += 1
    win_probability += [wins/(i)]
#print(wins)
#print(win_probability)
```

```python
win_probability2 = []
wins = 0
for i in range(1,10001):
    if monty(switch=True):
        wins += 1
    win_probability2 += [wins/(i)]
#print(wins)
#print(win_probability)
```
